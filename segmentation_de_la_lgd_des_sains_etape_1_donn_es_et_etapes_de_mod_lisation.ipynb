# **Bases de données pour la segmentation**


La base de données contient plus de 100 variables décrivant les informations démographiques et financières sur les entrées en défaut entre 2014 et 2022, et qui sont entrées en défaut selon la définition réglementaire imposée par l'Autorité Bancaire Européenne (EBA).

Les variables qui caractérisent les clients sains et qui seront utilisées pour la modélisation de la LGD Saine ont été récupérées un an avant l'entrée en défaut.

# **Projet de segmentation**

### *Contexte du Projet*

L'objectif d'une institution financière est d'anticiper la perte financière liée au risque. Pour le risque de crédit les institutions financières utilisant les modèles avancés doivent évaluer avec la plus grande rigueur la Probabilité de défaut (PD), la perte en cas de défaut (LGD) et l'Exposition au Défaut (EAD) en utilisant des modèles statistiques.

Ce projet utilise la base de données précédente pour modéliser la perte en cas de défaut des clients Sains. L'objectif est de développer un modèle de segmentation ayant pour but de classer les clients en classes homogènes de risque (CHR) selon l'axe du taux de perte.

### *Objectifs*

  
* Analyser et comprendre le profil des défauts.
* Identifier les variables clés influençant le taux de perte.
* Assurer la stabilité des répartitions et niveau de risque des variables.
* Construire et valider un modèle prédictif (ex. régression linéiare, arbres de régression, gradient boosting etc..).
* Assurer la stabilité et la robustesse du modèle et sa performance.
* Mettre en place des classes homogènes de risque (CHR)

### *Étapes Clés du Projet*

1. **Définition du Problème & Recueil des Exigences**
   - Comprendre le besoin métier/réglementaire.
   - Analyser l'existant
   - Analyser les exigences règlementaires
   - Définir les indicateurs de performance (AUC, RMSE, etc.) et fixer les objectifs.

2. **Collecte et Préparation des Données**
   - Points clés de la collecte
   - Importer le jeu de données.
   - Vérifier les exigences réglementaires
   - Nettoyer les données : gestion des valeurs manquantes, outliers, et incohérences.
   - Séparer le jeu de données en ensembles d'entraînement, de validation et de test.
   - Analyser la représentativité des données d'entrainement (variables pertinentes et indicateur de défaut)
   - Utiliser le jeu de données d'entrainement pour faire la suite du processus.

3. **Exploration des Données (EDA)**
   - Analyser les distributions des variables et visualiser les corrélations.


4. **Feature Engineering et Feature Selection**
   - Créer de nouvelles variables (ex. ratios, agrégations) pour enrichir le modèle.
   - Discrétisation des variables continues
   - Encodage des variables catégoriques (One-Hot, Target Encoding…)
   - Transformer et normaliser/standardiser les données.
   - Assurer la stabilité des répartitions et niveau de risque des variables.
  -  Analyser les corrélations et Identifier les variables pertinentes pour la prédiction.

5. **Modélisation**
   - Entraîner plusieurs modèles de classification(interprétabilité et explicabilité)
   - Optimiser les hyperparamètres par validation croisée.

6. **Évaluation du Modèle et segmenation**
   - Évaluer les performances globales et aussi au sein de sous populations matérielles(AUC, F1-score, etc.).
   - Analyser l'importance des variables.
   - Assurer la stabilité et la robustesse du modèle et sa performance.
   - Segmentation en classes homogènes de risque (nombre de classes)
   - Stabilité des classes homogènes de risque
   - Etude de l'homogénéité, de l'hétérogénéité et de la concentration des classes de risque (+ métiers si présence de classes métiers)

7. **Déploiement et Suivi**
   - Intégrer le modèle dans un environnement de production.
   - Mettre en place un suivi régulier et un système de recalibrage du modèle.
